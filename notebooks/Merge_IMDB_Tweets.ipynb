```python
import pandas as pd

file_path_imdb = "/content/imdb_sample_500.csv"
file_path_tweets = "/content/tweets_sample_500.csv"

df_imdb = pd.read_csv(file_path_imdb)
df_tweets = pd.read_csv(file_path_tweets)

```

```python
df_imdb.head()
```

```python
df_tweets.head()
```

```python
df_imdb = df_imdb.rename(columns={'review': 'text'})

df_imdb.head()
```

```python
df_tweets = df_tweets.rename(columns={'target': 'sentiment'})

df_tweets.head()
```

```python
df = pd.concat([df_imdb[['text', 'sentiment']], df_tweets[['text', 'sentiment']]])

df.head()
```

```python
print(df.value_counts())
```

```python
print(df.shape)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

df.head()
```

```python
from google.colab import files

df.to_csv('data.csv', index=False)
files.download('data.csv')
```
