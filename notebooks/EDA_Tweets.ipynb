```python
import pandas as pd

file_path="/tweets.csv"

with open(file_path, 'r') as f:
  for _ in range(5):
    print(f.readline())
```

```python
cols = ['target', 'ids', 'date', 'flag', 'user', 'text']

df = pd.read_csv(file_path, header=None, names=cols, encoding='latin-1')

df.head()
```

```python
print("Unique Target values: ", df['target'].unique())

print(df['target'].value_counts())
```

[Sentiment140 dataset](https://www.kaggle.com/datasets/kazanova/sentiment140/data/)

```python
df = df[['target', 'text']]

print(df.head())

print(df.shape)
```

```python
df['target'] = df['target'].replace(4,1)

print("unique target values: ", df['target'].unique())

print(df['target'].value_counts())
```

```python
df_neg = df[df['target'] == 0].sample(n=250, random_state=42)

df_pos = df[df['target'] == 1].sample(n=250, random_state=42)

df_sample = pd.concat([df_neg, df_pos])

df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)

print("New dataset shape: ", df_sample.shape)

print(df_sample['target'].value_counts())

print(df_sample.head())
```

```python
from google.colab import files

df_sample.to_csv('tweets_sample_500.csv', index=False)

files.download('tweets_sample_500.csv')
```
