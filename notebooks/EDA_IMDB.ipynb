```python
import pandas as pd

file_path = "/content/imdb.csv"

with open(file_path, 'r') as f:
  for _ in range(5):
    print(f.readline())
```

```python
df = pd.read_csv(file_path)

df.head()
```

```python
print("unique values of sentiment: ", df['sentiment'].unique())
```

```python
print("value count of sentiment: ", df['sentiment'].value_counts())
```

```python
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

print("unique values: ", df['sentiment'].unique() )
```

```python
print("count unique value: ", df['sentiment'].value_counts())
```

```python
df_neg = df[df['sentiment'] == 0].sample(n=250, random_state=42)
df_pos = df[df['sentiment'] == 1].sample(n=250, random_state=42)

df_sample = pd.concat([df_neg, df_pos])
```

```python
df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)
```
